```python
#importando bibliotecas
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity= "all"
```

```python
#importando csv
df = pd.read_csv('amazon.csv',encoding='latin-1')
```

```python
#Visualizando os primeiros 5 dados do dataset
df.head()
#Visualizando os últimos 5 dados do dataset
df.tail()
```

```python
df.info()
print(f'O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas')
```

```python
#Utilizando método loc para retornar da linha 100 até a 120
df.loc[100:120, ['state', 'month', 'number']]
#Utilizando método iloc para retornar da linha 100 até a 120
df.iloc[100:121, 1:4]
```

```python
#total de queimadas em 2010
df.loc[df['year'] == 2010, 'number'].sum()
#total de queimadas em 2010 no Mato Grosso do Sul
df.loc[(df['year'] == 2010) & (df['state'] == 'Mato Grosso'), 'number'].sum()
```

```python
#Média de incêndios no Acre por mês
df.loc[df['state'] == "Acre"].groupby('month')['number'].mean()
#Desvio Padrão de incêndios no Acre por mês
df.loc[df['state'] == "Acre"].groupby('month')['number'].std()
```

```python
#Média de incêndios por mês
df.groupby(['month', 'state'])['number'].mean()
#Maior média
df.groupby(['month', 'state'])['number'].mean().groupby( 'month').idxmax()
#Menor média
df.groupby(['month', 'state'])['number'].mean().groupby('month').idxmin()

```
