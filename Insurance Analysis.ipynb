```python
#usando np.select eu consegui criar uma tabela que se baseasse em 3 condições para o bmi
conditions = [
    (df["sex"] == "male") & (df["bmi"]<20),
    (df["sex"] == "male") & (df["bmi"]>=20) & (df["bmi"]<= 24.9),
    (df["sex"] == "male") & (df["bmi"]>24.9)
]
choices = ["Abaixo do peso", "Normal", "Acima do peso"]
df["bmiMale"] = np.select(conditions, choices, default = '')

conditions = [
    (df["sex"] == "female") & (df["bmi"]<18.5),
    (df["sex"] == "female") & (df["bmi"]>=18.5) & (df["bmi"]<= 25),
    (df["sex"] == "female") & (df["bmi"]>25)
]
choices = ["Abaixo do peso", "Normal", "Acima do peso"]
df["bmiFemale"] = np.select(conditions, choices, default = '')
df.head(10)



```
