Separando os 5 melhores times de 2019 presentes no fifa 19 por continente


```python
# Times da Europa
times_europa = [
    "Real Madrid",
    "FC Barcelona",
    "Manchester City",
    "Juventus",
    "Chelsea"
]

# Times da América do Sul
times_america_do_sul = [
    "Boca Juniors",
    "River Plate",
    "Santos",
    "Internacional",
    "Cruzeiro"
]

# Times da América do Norte
times_america_do_norte = [
    "LA Galaxy",
    "New York City FC",
    "Inter miami",
    "Atlanta United",
    "Seattle Sounders FC"
]

# Times da Ásia
times_asia = [
    "Guangzhou Evergrande Taobao FC",
    "Al Hilal",
    "Vissel Kobe",
    "Jeonbuk Hyundai Motors",
    "Urawa Red Diamonds"
]

# Times da Oceania
times_oceania = [
    "Melbourne Victory",
    "Sydney FC",
    "Wellington Phoenix",
    "Melbourne City FC",
    "Adelaide United"
]

# Times da África
times_africa = [
    "Kaizer Chiefs",
    "Orlando Pirates",
    "Al Ahli",
    "Al Ittihad",
    "Al Shabab"
]
```

Top 5 de cada continente com maiores médias de idade


```python
#Europa
europa = df[df['Club'].isin(times_europa)]
mean_europa_age = europa.groupby('Club')['Age'].mean().nlargest(5)
mean_europa_age

#América do Sul
america_do_sul = df[df['Club'].isin(times_america_do_sul)]
mean_america_do_sul_age = america_do_sul.groupby('Club')['Age'].mean().nlargest(5)
mean_america_do_sul_age

#América do Norte
america_do_norte = df[df['Club'].isin(times_america_do_norte)]
mean_america_do_norte_age = america_do_norte.groupby('Club')['Age'].mean().nlargest(5)
mean_america_do_norte_age

#Ásia
asia = df[df['Club'].isin(times_asia)]
mean_asia_age = asia.groupby('Club')['Age'].mean().nlargest(5)
mean_asia_age

#Oceania
oceania = df[df['Club'].isin(times_oceania)]
mean_oceania_age = oceania.groupby('Club')['Age'].mean().nlargest(5)
mean_oceania_age

#África
africa = df[df['Club'].isin(times_africa)]
mean_africa_age = africa.groupby('Club')['Age'].mean().nlargest(5)
mean_africa_age

```

# Análise Exploratória 

```python
import matplotlib.pyplot as plt
```

Criando um histograma para mostrar a frequência de overalls maiores ou iguais a 90

```python
plt.figure(figsize=(8,6))
plt.title("Top Overall Histogram")
df[df['Overall'] >= 90]['Overall'].hist(bins=10)
plt.xlabel('Overall')
plt.ylabel('Frequência')
plt.show();
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Skill Moves'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Skill moves')
plt.xlabel('Classificação')
plt.ylabel('Frequência')
plt.show();
```

Jogadores do Brasil com melhores skills

```python
top_brazilian_players= brazilian_players.nlargest(5, 'Skill Moves')
top_brazilian_players.hist(['Age'], ['Overall'], figsize=(15, 10), bins=20, xlabelsize=6, ylabelsize=6);
```

```python
avg_overall_by_age = df.groupby('Age')['Overall'].mean()
plt.figure(figsize=(10, 6))
avg_overall_by_age.plot(kind='line', color='green', marker='o')
plt.title('Média do overall por Idade')
plt.xlabel('Idade')
plt.ylabel('Média do overall')
plt.grid()
plt.show();
```
